## Librerías

```python
import spacy
```

## Ejercicio 1.1

### English

```python
nlp = spacy.load("en_core_web_sm")

with open("en.txt") as file:
    text = file.read()

doc = nlp(text)

for token in doc:
    print(token.text, token.pos_, token.tag_)
```

### Spanish

```python
nlp = spacy.load("es_core_news_sm")

with open("es.txt") as file:
    text = file.read()

doc = nlp(text)

for token in doc:
    print(token.text, token.pos_, token.tag_)
```
